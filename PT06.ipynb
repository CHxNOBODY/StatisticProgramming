```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
```

6.1 ร้านค้าออนไลน์ขายเมล็ดพืชหายากชนิดหนึ่ง อ้างว่าเมล็ดที่ขายมีเปอร์เซ็นต์การงอกสูงกว่า 90% จากการทดลองของลูกค้ารายหนึ่งได้สั่งซื้อเมล็ดมาทั้งหมด 100 เมล็ดและพบว่ามี 14 เมล็ดที่ไม่งอก จงทดสอบคำกล่าวอ้างของร้านค้าว่าเป็นจริงหรือไม่ที่ระดับนัยสำคัญ 0.05 หากถือว่าอัตราการงอกของเมล็ดพืชดังกล่าวมีการแจกแจงแบบปกติ

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล

# สมมติฐาน
# H0: P <= P0
# H1: P > P0

n = 100
alpha = 0.05

p_0 = 0.9
p_hat = 0.14

z_cal = (p_hat - p_0) / np.sqrt(p_0 * (1 - p_0) / n)
z_crit = stats.norm.isf(alpha)
p_value = stats.norm.sf(z_cal)

print(f"ค่า z_cal: {z_cal:.4f}")
print(f"ค่า z-crit: {z_crit:.4f}")
print(f"ค่า p-value: {p_value:.4f}")

if p_value < alpha:
    print("ปฏิเสธสมมติฐาน H0 : อ้างของร้านค้าว่าเป็นไม่จริง")
else:
    print("ยอมรับสมมติฐาน H0 : อ้างของร้านค้าว่าเป็นจริง")
```

6.2 สืบเนื่องจากข้อ 8.1 พบว่ามีร้านค้าออนไลน์อีกเจ้าซึ่งขายเมล็ดพืชหายากชนิดเดียวกันนี้ อ้างว่าเมล็ดที่ขายมีเปอร์เซนต์การงอกสูงกว่า 95% จึงสั่งซื้อมา 50 เมล็ดพบว่ามี 4 เมล็ดที่ไม่งอก จงทดสอบว่าร้านค้านี้ขายเมล็ดซึ่งมีเปอร์เซ็นต์การงอกสูงกว่าร้านค้าแรกมากกว่าหรือเท่ากับ 5% หรือไม่ ที่ระดับนัยสำคัญ 0.05

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล

# สมมติฐาน
# H0: P1 - P2 <= 0.05
# H1: P1 - P2 > 0.05

P1 = 4/50
P2 = 14/100

alpha = 0.05

z_cal = abs(P1 - P2) / np.sqrt((P1 * (1 - P1) / 50) + (P2 * (1 - P2) / 100))
z_crit = stats.norm.isf(alpha)
p_value = stats.norm.sf(z_cal)

print(f"ค่า z_cal: {z_cal:.4f}")
print(f"ค่า z-crit: {z_crit:.4f}")
print(f"ค่า p-value: {p_value:.4f}")

if p_value < alpha:
    print("ปฏิเสธสมมติฐาน H0 : ร้านค้านี้ขายเมล็ดซึ่งมีเปอร์เซ็นต์การงอกน้อยกว่าหรือเท่ากับร้านค้าแรก")
else:
    print("ยอมรับสมมติฐาน H0 : ร้านค้านี้ขายเมล็ดซึ่งมีเปอร์เซ็นต์การงอกสูงกว่าร้านค้าแรก")
```

6.3 จากชุดข้อมูล "titanic.csv" จงทดสอบว่าส่วนเบี่ยงเบนมาตราฐานข้อมูลอายุผู้โดยสารมีค่าน้อยกว่า 10 ปีหรือไม่ที่ระดับความเชื่อมั่น 99% หากกำหนดให้ข้อมูลอายุผู้โดยสารมีการแจกแจงแบบปกติ

```python
df = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")
df.head()
```

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล

# สมมติฐาน
# H0: std^2 >= (std_0) ^2
# H1: std^2 < (std_0) ^2

alpha = 0.01
ages = df['Age'].dropna()
n = len(ages)
std = ages.std()
std_0 = 10

x_squre_cal = (n-1)* (std**2) / (std_0**2)
x_squre_crit = stats.chi2.ppf(alpha, df = n-1)
p_value = stats.chi2.cdf(x_squre_cal, df = n-1)

print(f"ค่า x_squre_cal: {x_squre_cal:.4f}")
print(f"ค่า x_squre_crit: {x_squre_crit:.4f}")
print(f"ค่า p_value: {p_value:.4f}")

if p_value < alpha:
    print("ปฏิเสธสมมติฐาน H0 : ส่วนเบี่ยงเบนมาตราฐานข้อมูลอายุผู้โดยสารมีน้อยกว่า 10 ปี")
else:
    print("ยอมรับสมมติฐาน H0 : ส่วนเบี่ยงเบนมาตราฐานข้อมูลอายุผู้โดยสารมีค่ามากกว่าหรือเท่ากับ 10 ปี")
```

6.4 จากข้อ 6.3 จงทดสอบว่ากลุ่มผู้โดยสารชั้น Pclass = 1 มีความแปรปรวนของอายุเท่ากับ กลุ่มผู้โดยสารชั้น Pclass = 2 หรือไม่ที่ระดับนัยสำคัญ 0.05

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล

# สมมติฐาน
# H0: std_1^2 =  std_2^2
# H1: std_1^2 !=  std_2^2

ages_1 = df[df['Pclass'] == 1]['Age'].dropna()
ages_2 = df[df['Pclass'] == 2]['Age'].dropna()

n_1 = len(ages_1)
n_2 = len(ages_2)

var_1 = ages_1.var()
var_2 = ages_2.var()

print(var_1)
print(var_2)

F_cal = var_1 / var_2
df_1 = n_1 - 1
df_2 = n_2 - 1

alpha = 0.05
p_value = stats.f.sf(F_cal, df_1, df_2)
F_crit_1 = stats.f.ppf(1 - alpha / 2, df_1, df_2)
F_crit_2 = stats.f.ppf(alpha / 2, df_1, df_2)

print(f"ค่า F_cal: {F_cal:.4f}")
print(f"ค่า p_value: {p_value:.4f}")
print(f"ค่า F_crit_1: {F_crit_1:.4f}")
print(f"ค่า F_crit_2: {F_crit_2:.4f}")

if p_value < alpha:
    print("ปฏิเสธ H0: ความแปรปรวนของอายุผู้โดยสารชั้น 1 และชั้น 2 ไม่เท่ากัน")
else:
    print("ยอมรับ H0: ไม่มีหลักฐานเพียงพอว่าความแปรปรวนของอายุผู้โดยสารชั้น 1 และชั้น 2 แตกต่างกัน")
```
